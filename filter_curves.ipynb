```python
import matplotlib.pylab as plt 
import matplotlib.lines as mlines
from matplotlib.legend import Legend
```

```python
data_decam_g = open('/Users/mtownsend/anaconda/Data/decam_g.txt', 'r')
data_decam_r = open('/Users/mtownsend/anaconda/Data/decam_r.txt', 'r')
data_decam_z = open('/Users/mtownsend/anaconda/Data/decam_z.txt', 'r')

data_HSC_g = open('/Users/mtownsend/anaconda/Data/HSC_g.txt', 'r')
data_HSC_r = open('/Users/mtownsend/anaconda/Data/HSC_r.txt', 'r')
data_HSC_z = open('/Users/mtownsend/anaconda/Data/HSC_z.txt', 'r')

data_90Prime_g = open('/Users/mtownsend/anaconda/Data/90Prime_g.txt', 'r')
data_90Prime_r = open('/Users/mtownsend/anaconda/Data/90Prime_r.txt', 'r')
data_MOSAIC_z = open('/Users/mtownsend/anaconda/Data/MOSAIC_z.txt', 'r')


lines_decam_g = data_decam_g.readlines()[1:]
lines_decam_r = data_decam_r.readlines()[1:]
lines_decam_z = data_decam_z.readlines()[1:]

lines_HSC_g = data_HSC_g.readlines()[1:]
lines_HSC_r = data_HSC_r.readlines()[1:]
lines_HSC_z = data_HSC_z.readlines()[1:]

lines_90Prime_g = data_90Prime_g.readlines()[5:]
lines_90Prime_r = data_90Prime_r.readlines()[5:]
lines_MOSAIC_z = data_MOSAIC_z.readlines()[2:]
```

```python

decam_lambda_g = []
decam_thru_g = []

for line in lines_decam_g:
    p = line.split()
    decam_lambda_g.append(float(p[0]))
    decam_thru_g.append(float(p[1]))
    
decam_lambda_r = []
decam_thru_r = []

for line in lines_decam_r:
    p = line.split()
    decam_lambda_r.append(float(p[0]))
    decam_thru_r.append(float(p[1]))
    
decam_lambda_z = []
decam_thru_z = []

for line in lines_decam_z:
    p = line.split()
    decam_lambda_z.append(float(p[0]))
    decam_thru_z.append(float(p[1]))
    
    
HSC_lambda_g = []
HSC_thru_g = []

for line in lines_HSC_g:
    p = line.split()
    HSC_lambda_g.append(float(p[0]))
    HSC_thru_g.append(float(p[1]))
    
HSC_lambda_r = []
HSC_thru_r = []

for line in lines_HSC_r:
    p = line.split()
    HSC_lambda_r.append(float(p[0]))
    HSC_thru_r.append(float(p[1]))
    
HSC_lambda_z = []
HSC_thru_z = []

for line in lines_HSC_z:
    p = line.split()
    HSC_lambda_z.append(float(p[0]))
    HSC_thru_z.append(float(p[1]))
    

Prime_lambda_g = []
Prime_thru_g = []

for line in lines_90Prime_g:
    p = line.split()
    Prime_lambda_g.append(float(p[0]))
    Prime_thru_g.append(float(p[3]))
    
Prime_lambda_r = []
Prime_thru_r = []

for line in lines_90Prime_r:
    p = line.split()
    Prime_lambda_r.append(float(p[0]))
    Prime_thru_r.append(float(p[3]))
    
MOSAIC_lambda_z = []
MOSAIC_thru_z = []

for line in lines_MOSAIC_z:
    p = line.split()
    MOSAIC_lambda_z.append(float(p[0]))
    MOSAIC_thru_z.append(float(p[1]))
    
    
```

```python
plt.figure(figsize=(10,10))
plt.plot(decam_lambda_g, decam_thru_g, c='green', label='DECam g')
plt.plot(decam_lambda_r, decam_thru_r, c='red', label='DECam r')
plt.plot(decam_lambda_z, decam_thru_z, c='blue', label='DECam z')
plt.plot(HSC_lambda_g, HSC_thru_g, c='springgreen', label='HSC g')
plt.plot(HSC_lambda_r, HSC_thru_r, c='lightcoral', label='HSC r')
plt.plot(HSC_lambda_z, HSC_thru_z, c='deepskyblue', label='HSC z')
plt.legend(loc='upper right', fontsize=15)
plt.savefig('/Users/mtownsend/Desktop/MeetingPlots/filter_curves_decam.pdf')
plt.show()

plt.figure(figsize=(10,10))
plt.plot(Prime_lambda_g, Prime_thru_g, c='green', label='90Prime g')
plt.plot(Prime_lambda_r, Prime_thru_r, c='red', label='90Prime r')
plt.plot(MOSAIC_lambda_z, MOSAIC_thru_z, c='blue', label='MOSAIC z')
plt.plot(HSC_lambda_g, HSC_thru_g, c='springgreen', label='HSC g')
plt.plot(HSC_lambda_r, HSC_thru_r, c='lightcoral', label='HSC r')
plt.plot(HSC_lambda_z, HSC_thru_z, c='deepskyblue', label='HSC z')
plt.legend(loc='upper right', fontsize=15)
plt.savefig('/Users/mtownsend/Desktop/MeetingPlots/filter_curves_bass_mzls.pdf')
plt.show()
```
